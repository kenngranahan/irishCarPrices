```python
import json
from pathlib import Path
from tqdm import tqdm
```

```python
path = Path('scraped_data/')
error_log = 'cleaned_data/error_log.txt'
ids_cleaned = []


for file in tqdm(list(path.glob('*'))):
    id_status = {'Good Download': False, 
                 'Checked Attributes': False}

    file_name = file.name
    file_id = file_name.split('.')[0]
    file_id = file_id[4:]
    with open(file, 'r') as f:
        data = json.loads(f.read())

    try:
        id_status['Good Download'] = check_good_download(data)    
    except IndexError:
        with open(error_log, 'a') as f:
            f.write('Good Download:'+ file_id + '\n')

    if id_status['Good Download']:
        try:
            data = check_attributes(data)
            id_status['Checked Attributes'] = True
        except IndexError:
            with open(error_log, 'a') as f:
                f.write('Checked Attributes:'+ file_id + '\n')

        if id_status['Checked Attributes']:
            with open('cleaned_data/'+file_name, 'w') as f:
                f.write(json.dumps(data))
```

```python
def check_good_download(data):
    if data['dealer'][0][:3] == '<!D':
        return False
    return True

def check_attributes(data):
    types = ['Manual', 'Automatic' ,'Other' ,'CVT', 'Semi-auto']
    
    fuel = ['Petrol', 'Diesel', 'Petrol Hybrid', 'Electric', 'Hybrid', 'Other', 'Bio gas'
            'Petrol Plug-in Hybrid', 'Diesel Hybrid', 'Diesel Plug-in Hybrid']
    
    color = ['Black', 'Grey', 'White', 'Blue', 'Silver', 'Red', 'Brown', 'Bronze', 'Green', 'Orange']
    
    body = ['Hatchback', 'Saloon', 'SUV', 'MPV', 'Estate', 'Other', 'Van', 'Coupe', 'Pickup', 'Convertible']
    
    
    def is_type(text):
        return (text in types)
    
    def is_engine(text):
        return (text[-1] == 'L' or text == 'Unknown')
    
    def is_body(text):
        return (text in body)
    
    def is_fuel(text):
        return (text in fuel)
    
    def is_doors(text):
        return (len(text)>5 and (text[-5:] == 'Doors' or text[-4:] == 'Door'))
    
    def is_mpg(text):
        return (len(text)>3 and text[-3:]=='mpg')
    
    def is_owners(text):
        return (len(text) >=7 and (text[-6:] == 'Owners' or text[-5:]=='Owner'))
    
    def is_colors(text):
        return (text in color)
    
    def is_tax(text):
        return (len(text)>0 and text[0] == '€')
    
    def is_nct(text):
        return (len(text) > 3 and text[:3] == 'NCT')
    
    identify_attribute = {'type': is_type,
                          'engine': is_engine,
                          'body': is_body,
                          'fuel': is_fuel,
                          'doors':is_doors,
                          'mpg':is_mpg,
                          'owners':is_owners,
                          'color':is_colors,
                          'tax':is_tax,
                          'nct':is_nct,}
    correct_attributes = {}
    
    for attribute in identify_attribute:
        text = data[attribute][0]
        
        if not identify_attribute[attribute](text):
            for correct_attribute in identify_attribute:
                if identify_attribute[correct_attribute](text):
                    correct_attributes[correct_attribute] = text
            data[attribute][0] = ''
                    
    if len(correct_attributes) != 0:
        for attribute in correct_attributes:
            data[attribute][0] = correct_attributes[attribute]

    return data
```

```python
name = 'ConnollyVolkswagenLetterkenny'
def spilt_name(data):
    name = data['dealer'][0]
    new_name = ''
    for char in name:
        if char.isupper():
            new_name = new_name + ' '
        new_name = new_name + char
    return new_name[1:]



def split_vehicle_name(data):
    vehicle = data['vehicle'][0]
    new_vehicle = ''
    new_vehicle = new_vehicle + vehicle[0]
    for idx, char in enumerate(vehicle[1:]):
        if char.isupper():
            if vehicle[idx+2].isupper():
                new_vehicle = new_vehicle + char.lower()
            else:
                new_vehicle = new_vehicle + ' '
                new_vehicle = new_vehicle + char
        else:
            new_vehicle = new_vehicle + char
    return new_vehicle
```

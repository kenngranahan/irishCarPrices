```python
sns.displot(data=carData, x='price')
```

```python
from math import log
carData['log_price'] = carData['price'].apply(lambda price:log(price))
sns.displot(data=carData, x='log_price')
```

#### Now let's look at the correlation matrix to get a sense of which features have the most predictive power. It looks like mileage, vehicle_year and engine size all have high correlations with the price, but what's more interesting is that the have an even better correalation with log_price

```python
sns.heatmap(carData.corr(), cmap='coolwarm')
```

Given that it's normally distributed and linear in these feature, some linear regression model would be a good place to start

#### To remove the outliers in vehicle_year I think we can simply drop any sample older than 25 years. The rationale for this is that once a vehicle is older than 25 years it is likely a vintage model or it's getting sold for scrap metal. I think the economics driving the price in these scenarios are very different from what sets prices in the rest of the second hand market, so we'll just drop these from the dataset

```python
from datetime import date
carData['age'] = (date.today().year - carData['vehicle_year'])
carData = carData[carData['age']<25]
```

#### The outliers in the mileage variable are caused by people deliberatly understating the mileage on the car when the vehicle is listed for sale. We can see this by checking how many previous owners a car has had. As we can see there are 260 cars with 0 mileage but have had previous owners.

```python
carsWithNoMileage = carData[carData['mileage']==0]
carsWithNoMileage.groupby('owners').count()['model']
```

#### We see from the Ford Focus example that if we plot miles_per_year against age we have 3 vehicles which are 7,11 and 13 years old with close to 0 kilometres driven. This is clealy and artefact in the data

```python
fig, ax = plt.subplots(figsize=(15,10))

carData['miles_per_year'] = carData['mileage']/carData['age']
dataToPlot = carData[(carData['make']=='Ford') & (carData['model']=='Focus')]
sns.scatterplot(data=dataToPlot, x='age', y='miles_per_year', ax=ax)

plt.show()
```

#### I think we can identify outliers in mileage by calculating mileage/age, i.e. the number of killometres driven per year. For cars older than 1 year, if this variable is below a particular threshold we'll tag the corresponding mileage as an outlier. We see below that if I introduce 1000 kilometres per year as the cut off and run this logic it correctly tags these 3 data points as outliers

```python
fig, ax = plt.subplots(figsize=(15,10))


carData['mileage_is_ourlier'] = ((carData['miles_per_year']<1000) & (carData['age']>1))
dataToPlot = carData[(carData['make']=='Ford') & (carData['model']=='Focus')]
sns.scatterplot(data=dataToPlot, x='age', y='miles_per_year', ax=ax, hue='mileage_is_ourlier')

plt.show()
```

#### We can replace these outliers by estimating the true mileage of the vehicle has on the national average. According to the CSO, Irish drivers drive, on average, 17000km every year. So we can estimate the mileage as 17000 * age. I'll also add some gaussian noise to the calculation as I don't want some samples of mileage and age to have a 1-1 mapping. The standard deviation of the noise will be equal to the standard deviation of the mileage variable

```python
from numpy import random
rng = random.default_rng()

mileageStd = carData[carData['mileage']>0]['mileage'].std()
noise = rng.normal(loc=17000,scale=mileageStd, size=len(carData))

carData['mileage_fill_values'] =  (carData['age']-1)*17000 + noise
carData['replace_mileage'] = carData['mileage']/carData['age']
carData['replace_mileage'] = ((carData['replace_mileage'] < 1000) & (carData['age']>1))

carData['mileage'] = carData['mileage'].where(~carData['replace_mileage'], carData['mileage_fill_values'])
    
```

#### Last thing to check is the missingness of the dataset

```python
carData.isna().sum()
```

#### The missingness in the price data is due to some cars being listed as POA. I think we can drop these samples as there simply isn't a target variable for these. For the remaining missing values I'll fill them using the median since it's only numeric columns missing values. I'll use the median of the car's model to fill it, if the median doesn't exist I'll use the car's make and if that alos doesn't exist I'll use the median of the overall dataset

```python
carData = carData[~carData['price'].isna()]

toFill = ['engine', 'doors', 'mpg', 'owners', 'tax', 'nct_year']


medianModelData = carData.groupby(by=['make','model'])[toFill ].median()
medianMakeData = carData.groupby(by='make')[toFill ].median()
medianData = carData[toFill].median()

carData=carData.merge(medianModelData[toFill], how='left', 
                      left_on=['make', 'model'], right_index=True)

for feature in toFill:
    carData[feature] =  carData[feature+'_x'].fillna(carData[feature+'_y'])
    carData.drop(labels = [feature+'_x', feature+'_y'], axis=1, inplace=True)

carData=carData.merge(medianMakeData[toFill ], how='left', 
                      left_on='make', right_index=True)
for feature in toFill :
    carData[feature] =  carData[feature+'_x'].fillna(carData[feature+'_y'])
    carData.drop(labels = [feature+'_x', feature+'_y'], axis=1, inplace=True)    

carData.fillna(medianData, inplace=True)
```
